## Plot the Loss

We will now create a 4x4 plot of the training/validation loss graphs (including the variants with light tuning of learning rate). Code snippet used for labeling the columns/rows extracted and modified from: https://exceptionshub.com/row-and-column-headers-in-matplotlibs-subplots.html

```python
# ================================================================================
# FUNCTION: populate_subplot( ax , row , col , data , tuned_data )
# ================================================================================
# 
# Input:
#   - ax: the set of matplotlib axes on which the subplot resides
#   - row: the row on which subplot is to be populated (0 indexed)
#   - col: the column on which subplot is to be populated (0 indexed)
#   - data: the csv history table as a pandas dataframe for the un-tuned variant
#   - tuned_data: the csv history table as a pandas dataframe for the tuned variant
#
# Output:
#   - ax is modified to contain the appropriate plots in the specified location
#   - list containing all of the plots objects added to the figure
#        * in the order: training loss, validation loss, tuned training loss, 
#                        tuned validation loss
#
# Task:
#   - populate the subplot in the given location of the figure with four loss plots
#   - return the plots generated
#
# ================================================================================
def populate_loss_subplot(ax, row, col, data, tuned_data, training_label, validation_label):
    x       = range( len( data['Epoch'] ) )
    y_train = data[training_label]
    y_val   = data[validation_label]
    y_tune  = tuned_data[training_label]
    y_tune2 = tuned_data[validation_label]
    l1      = ax[row,col].plot(x,y_train,c='cornflowerblue')
    l2      = ax[row,col].plot(x,y_val,c='orange')
    l3      = ax[row,col].plot(x,y_tune,c='black',alpha=0.5)
    l4      = ax[row,col].plot(x,y_tune2,c='gray',alpha=0.4)
    return [ l1[0] , l2[0] , l3[0] , l4[0] ]
```

```python
# === Prepare the Subplots === #
plt.figure( figsize=(10,6) )
fig , ax = plt.subplots( nrows=4 , ncols=4 , figsize=(11,8) )
plt.setp( ax.flat , xlabel='Epoch' , ylabel='Loss' )
plt.tight_layout( )



# === Label the Rows and Columns === #
columns = [ 'No Preprocessing' , 'Small Grayscale Regularized' , 'PCA' , 'LSH' ]
rows    = [ 'Cat vs Dog:\nCustom Model\n\n' , 'Cat vs Dog:\nVGG19\n\n' , 'Intel Images:\nCustom Model\n\n' , 'Intel Images:\nVGG19\n\n' ]
pad     = 0.4
for a, col in zip(ax[0], columns):
    a.annotate(col, xy=(0.5, 1.12), xytext=(0, pad),
                xycoords='axes fraction', textcoords='offset points',
                size='large', ha='center', va='baseline')
for a, row in zip(ax[:,0], rows):
    a.annotate(row, xy=(-0, 0.5), xytext=(-a.yaxis.labelpad - pad, 0),
                xycoords=a.yaxis.label, textcoords='offset points',
                size='large', ha='center', va='center', rotation=90 )
    

    
# === Plot Row 0: Cat vs Dog: Custom Model === #
lines = populate_loss_subplot(ax, 0, 0, custom_catvdog_nopreprocess_history, tuned_custom_catvdog_nopreprocess_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 0, 1, custom_catvdog_smallgrayreg_history, tuned_custom_catvdog_smallgrayreg_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 0, 2, custom_catvdog_pca_history, tuned_custom_catvdog_pca_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 0, 3, custom_catvdog_lsh_history, tuned_custom_catvdog_lsh_history, 'Training Loss', 'Validation Loss')



# === Plot Row 1: Cat vs Dog: VGG19  === #
lines = populate_loss_subplot(ax, 1, 0, vgg19_catvdog_nopreprocess_history, tuned_vgg19_catvdog_nopreprocess_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 1, 1, vgg19_catvdog_smallgrayreg_history, tuned_vgg19_catvdog_smallgrayreg_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 1, 2, vgg19_catvdog_pca_history, tuned_vgg19_catvdog_pca_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 1, 3, vgg19_catvdog_lsh_history, tuned_vgg19_catvdog_lsh_history, 'Training Loss', 'Validation Loss')



# === Plot Row 2: Intel Images: Custom Model === #
lines = populate_loss_subplot(ax, 2, 0, custom_intel_nopreprocess_history, tuned_custom_intel_nopreprocess_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 2, 1, custom_intel_smallgrayreg_history, tuned_custom_intel_smallgrayreg_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 2, 2, custom_intel_pca_history, tuned_custom_intel_pca_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 2, 3, custom_intel_lsh_history, tuned_custom_intel_lsh_history, 'Training Loss', 'Validation Loss')



# === Plot Row 3: Cat vs Dog: VGG19  === #
lines = populate_loss_subplot(ax, 3, 0, vgg19_intel_nopreprocess_history, tuned_vgg19_intel_nopreprocess_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 3, 1, vgg19_intel_smallgrayreg_history, tuned_vgg19_intel_smallgrayreg_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 3, 2, vgg19_intel_pca_history, tuned_vgg19_intel_pca_history, 'Training Loss', 'Validation Loss')
lines = populate_loss_subplot(ax, 3, 3, vgg19_intel_lsh_history, tuned_vgg19_intel_lsh_history, 'Training Loss', 'Validation Loss')



# === Show the Subplots === #
fig.legend(lines,['Training Loss', 'Validation Loss', 'Tuned Training Loss', 'Tuned Validation Loss'])
fig.suptitle( 'Experiment Loss Results' , y=1.13 , fontsize=18 )
plt.show( )
```

## Construct Table of Model History Stats

Now that we've shown the graphs of the overall trends of the training/validation accuracies and losses for each of the 16 experiments (both tuned and no tuning), let's also construct a table which can be simple for readers to understand, demonstrating the "best" value achieved for each of the plotted metrics for each experiment. Note that the experiments were tuned by running the `kerastuner` package to test 10 different learning rates and use the one which converged to the best validation accuracy after 10 epochs (that learning rate was then used to train the model from scratch for 25 epochs).

```python
# ================================================================================
# FUNCTION: add_history_data_entry( arr , dataset , model , method , history , tuned )
# ================================================================================
#
# Input: 
#   - arr: the array to be updated with the new entry
#   - dataset: string label for the dataset of the entry
#   - model: string label for the dataset of the entry
#   - method: string label for the preprocessing method of the entry
#   - history: pandas dataframe containing the training history of the experiment
#   - tuned: pandas dataframe containing the tuning history results of the experiment
#
# Output:
#   - No return values
#   - arr is modified with a new entry appended
#
# Task:
#   - Construct the new entry with proper formatting
#   - append the entry to the given array
#
# ================================================================================
def add_history_data_entry( arr , dataset , model , method , history , tuned ):
    entry = {
        'Dataset'                       : dataset, 
        'Model'                         : model, 
        'Method'                        : method, 
        'Max Training Accuracy'         : history['Training Accuracy'].max( ),
        'Max Validation Accuracy'       : history['Validation Accuracy'].max( ),
        'Min Training Loss'             : history['Training Loss'].min( ),
        'Min Validation Loss'           : history['Validation Loss'].min( ),
        'Max Tuned Training Accuracy'   : tuned['Training Accuracy'].max( ),
        'Max Tuned Validation Accuracy' : tuned['Validation Accuracy'].max( ),
        'Min Tuned Training Loss'       : tuned['Training Loss'].min( ),
        'Min Tuned Validation Loss'     : tuned['Validation Loss'].min( )
    }
    arr.append( entry )
    return
```

## Construct Table of Model Time Stats

Now that we've shown the other graphs and tables necessary, let's examine the overhead cost in terms of seconds of runtime for all preprocessing steps (including tuning time). 

```python
# ================================================================================
# FUNCTION: add_time_data_entry( arr , dataset , model , method , tuning , data )
# ================================================================================
#
# Input:
#   - arr: list to be updated
#   - dataset: string label for the dataset used in the experiment
#   - model: string label for the model architecture used in the experiment
#   - method: string label for the preprocessing method used for the experiment
#   - tuning: string label to indicate whether or not the model was tuned
#   - data: pandas dataframe containing the time data for the experiment
#
# Output:
#   - No return
#   - arr will be updated with a formatted dictionary entry appended
#
# Task:
#   - using the given data and labels, construct a table entry as a dict
#   - append the properly formatted dict to the end of the given list (arr)
#
# ================================================================================
def add_time_data_entry( arr , dataset , model , method , data ):
    
    # === Set up Baseline Entries === #
    entry = {
        'Dataset' : dataset,
        'Model'   : model,
        'Method'  : method
    }
    total = data['Time (sec)'].sum( )
    
    # === Extract Loading Times === #
    if len( data[data['Event'] == 'Load/Process Dataset'] ) > 0:
        process = data[data['Event'] == 'Load/Process Dataset']['Time (sec)'].tolist( )[0]
    else:
        process = float( 0 )
    
    
    # === Extract Training Times === #
    if len( data[data['Event'] == 'Training'] ) > 0:
        train = data[data['Event'] == 'Training']['Time (sec)'].tolist( )[0]
    else:
        train = float( 0 )
    
    
    # === Extract Hyperparameter Tuning Times === #
    if len( data[data['Event'] == 'Hyperparameter Tuning'] ) > 0:
        tune = data[data['Event'] == 'Hyperparameter Tuning']['Time (sec)'].tolist( )[0]
    else:
        tune = float( 0 )
    
    
    # === Add Extracted Data to Entry === #
    entry['Load/Process Dataset (Sec)']          = abs( round( process , 2 ) )
    entry['Method Preprocessing Overhead (Sec)'] = abs( round( total - process - train - tune, 2 ) )
    entry['Tuning Time (Sec)']                   = abs( round( tune , 2 ) )
    entry['Training (Sec)']                      = abs( round( train , 2 ) )
    entry['Total Time (Sec)']                    = abs( round( total , 2 ) )
    
    # === Update the Given List === #
    arr.append( entry )
    return
```
